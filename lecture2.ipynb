```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

diamonds = sns.load_dataset("diamonds")

display(diamonds.head())
print("Dataset Shape:", diamonds.shape)
print("Missing Values:\n", diamonds.isnull().sum())

plt.figure(figsize=(8,5))
sns.countplot(data=diamonds, x='cut', order=diamonds['cut'].value_counts().index, palette='viridis')
plt.title("Distribution of Diamond Cut")
plt.show()

plt.figure(figsize=(8,5))
sns.histplot(diamonds['carat'], bins=50, kde=True)
plt.title("Distribution of Carat")
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(data=diamonds, x='carat', y='price', alpha=0.5)
plt.title("Carat vs Price")
plt.show()

plt.figure(figsize=(8,5))
sns.boxplot(data=diamonds, x='cut', y='price', palette='coolwarm')
plt.title("Price by Cut")
plt.show()

corr_matrix = diamonds.corr(numeric_only=True)
plt.figure(figsize=(10,6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix of Diamond Attributes")
plt.show()

g = sns.FacetGrid(diamonds, col='cut', col_wrap=3, height=4)
g.map(sns.scatterplot, 'carat', 'price', alpha=0.5)
plt.show()

plt.figure(figsize=(8,5))
sns.scatterplot(data=diamonds, x='carat', y='price')
plt.xscale('log')
plt.yscale('log')
plt.title("Carat vs Price (Log Scale)")
plt.show()
```
