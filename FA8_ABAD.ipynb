```python
import numpy as np
import pandas as pd
from scipy.stats import skew, sem, kurtosis
df = pd.read_csv("https://raw.githubusercontent.com/Apress/data-analysis-and-visualization-using-python/master/Ch07/Salaries.csv")
df.head(10)
profsal = df.groupby('rank')[['salary']]
```

```python
profsal.describe()
```

```python
missingval = profsal.agg(lambda x: x.isna().sum())
missingval
# missing values are shown below
```

```python
profsal.median()
# median is shown below
```

```python
profsal.mean()
# mean is shown below
```

```python
validval = profsal.count()
validval
# valid values are shown below
```

```python
stderr_mean = profsal.std()/(profsal.count())**(1/2)
stderr_mean
# std. error of mean is shown below
```

```python
profsal.std()
# std. deviation is shown below
```

```python
kurtosis = df.groupby('rank')[['salary']].agg(kurtosis)
kurtosis
# kurtosis is shown to be slightly different from the values in the given file
```

```python
profsal.skew()
# skewness
```

```python
stderr_mean = profsal.std()/(profsal.count())**(1/2)
stderr_mean
# std. error of mean
```

```python
var_coef = profsal.std()/profsal.mean()
var_coef
# coefficient of variation
```

```python
n = profsal.count()
standerr_skew = np.sqrt((6*n*(n-1))/((n-2)*(n+1)*(n+3)))
standerr_skew
# std. error of skewness was obtained using the formula 
```

```python
for i in range(25,100,25):
    print(f"\n{i}th percentile")
    nth_percent = profsal.quantile(q=i/100)
    print(nth_percent)
    
for i in range(10,100,10):
    print(f"\n{i}th percentile")
    nth_percent = profsal.quantile(q=i/100)
    print(nth_percent)
```
